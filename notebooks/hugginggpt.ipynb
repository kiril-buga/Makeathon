```python
from transformers import load_tool
import accelerate
hf_tools = [
    load_tool(tool_name)
    for tool_name in [
        "document_question_answering",
        "image_captioner",
        "image_question_answering",
        "image_segmentation",
        "speech_to_text",
        "summarization",
        "text_classification",
        "text_question_answering",
        "translation",
        "huggingface_tools/text_to_image",
        "huggingface_tools/text_to_video",
        "text_to_speech",
        "huggingface_tools/text_download",
        "huggingface_tools/image_transformation",
    ]
]
```

```python
# Setup model and HuggingGPT
# We create an instance of HuggingGPT and use ChatGPT as the controller to rule the above tools.

from langchain_experimental.autonomous_agents import HuggingGPT
from langchain_openai import OpenAI
from Groq import Groq

# %env OPENAI_API_BASE=http://localhost:8000/v1
llm = OpenAI(model_name="gpt-3.5-turbo")
agent = HuggingGPT(llm, hf_tools)
```

```python
agent.run("please show me a video and an image of 'a boy is running'")
```
